```python
df = df.drop('Uniformity_of_cell_size', axis=1)
```

```python
df
```

```python
train_data = df.sample(frac=0.8, random_state=123)
test_data = df.drop(train_data.index)
```

```python
X_train = train_data.drop('Class', axis=1).to_numpy()
X_test = test_data.drop('Class', axis=1).to_numpy()
y_train = train_data['Class'].to_numpy()
y_test = test_data['Class'].to_numpy()
```

## -Modelo

```python
class KNN:
    def __init__(self, X_train, y_train, k):
        self.X_train = X_train
        self.y_train = y_train
        self.k = k

    def calc_distance(self, X_train, X_test):
        return np.linalg.norm(X_train - X_test)

    def k_neighbors(self, X_test):
        distances = []
        for i in range(len(self.X_train)):
            distances.append(self.calc_distance(self.X_train[i], X_test))
        sorted_index = np.argsort(distances)
        k_nearest = sorted_index[:self.k]
        return k_nearest

    def classification(self, X_test):
        predictions = []
        k_nearest = self.k_neighbors(X_test)
        for i in k_nearest:
            predictions.append(self.y_train[i])
        count = np.bincount(predictions)
        max = np.argmax(count)
        return max

    def accuracy(self, predictions, y_test):
        test = predictions == y_test
        p = (np.sum(test)/y_test.shape[0])*100
        return p
```

```python
K = [3, 5, 7, 9, 11]

for k in K:
    predictions = []
    classificador = KNN(X_train, y_train, k)
    for x in X_test:
        predictions.append(classificador.classification(x))
    accuracy = classificador.accuracy(predictions, y_test)
    print(f'K = {k}, accuracy = {accuracy}')
```
